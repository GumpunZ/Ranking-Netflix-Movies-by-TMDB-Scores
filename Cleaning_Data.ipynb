# Cleaning


```python
import pandas as pd
```

```python
movie = pd.read_csv('final_movie_data.csv')
movie.head(10)
```

```python
#คัดเอาหนังที่มี่จำนวนโหวตตั้งแต่ 10 คนขึ้นไป
MovieVote = movie.loc[movie.vote_count >= 10]
MovieVote.head()
```

```python
# เอาหนังที่ไม่ทราบประเภทออก
MovieVote.drop(MovieVote.loc[MovieVote['genres'] == '[]'].index, inplace=True)
```

```python
M = MovieVote.copy()
M.isnull().sum()
```

# แยก Entity

```python
# เราสามารถใช้ลิสต์ของ data dict ที่มี key เหมือน
genres = []
gen = [ eval(i) for i in  M['genres']] #เข้าถึงข้อมูลในคอลัมน์ 'genres' ใช้ eval เพิ่มอ่าน str ให้เป็น list
for i in gen: #เข้าถึงข้อมูลใน gen แต่ว่าจะเป็นลิสต์ซ้อนลิสต์
  for j in i: #เข้าถึงข้อมูลในลิสต์ซ้อนลิสต์
    for k in j: #เข้าถึงข้อมูลที่อยู่ในลิสต์ ซึ่งในที่นี้จะเป็น data dict
      genres.append(j)

df_gen = pd.DataFrame(genres)
df_gen.drop_duplicates(inplace=True)

print(df_gen.shape)
df_gen.head(10) #df ของประเภทหนัง
```

```python
countries = []
coun = [ eval(i) for i in  M['production_countries']]
for i in coun:
  for j in i:
    countries.append(j)

df_ct = pd.DataFrame(countries)
df_ct.drop_duplicates(inplace=True)

print(df_ct.shape)
df_ct.head(10)  #df ของประเทศ
```

```python
directors = []
for i in  M['director']: #เข้าถึงข้อมูลในลิสต์ซ้อนลิสต์
  if ',' in i:
    i = i.split(',')
    for j in i:
      directors.append(j)
    continue
  directors.append(i)

df_drt = pd.DataFrame(directors)
df_drt.drop_duplicates(inplace=True)
df_drt.reset_index(inplace=True)
df_drt.drop(columns=['index'], axis=1, inplace=True)
df_drt.reset_index(inplace=True)
df_drt.rename(columns={'index':'id', 0:'name'}, inplace=True)

print(df_drt.shape)
df_drt.head(10)  #df ของผู้กำกับ
```

```python
casts = []
col = [  'id', 'name', 'gender']
gender = { 0:'unknow', 1:'female', 2:'male', 3: 'LGBTQ+' }
ls_cast = [ eval(i.replace('false', 'False').replace('null', 'None').replace('true', 'True')) for i in M['cast']]
for i in ls_cast:
  for j in i:
    j['gender'] = gender[j['gender']]
    casts.append(j)

df_cast = pd.DataFrame(casts)
df_cast = df_cast[[c for c in col]]
df_cast.drop_duplicates(inplace=True)

print(df_cast.shape)
df_cast.head(10)  #df ของหนังแสดง
```

```python
col = ['id', 'title','description', 'release_year',
        'runtime',  'vote_average',
        'vote_count']

df_movie = M[col]

print(df_movie.shape)
df_movie.head(10) #df ของหนัง
```

# Relationship

```python
import re
```

```python
map = M[['id', 'genres', 'director', 'cast', 'production_countries']]
map.reset_index(inplace=True)
map.drop(columns=['index'], axis=1, inplace=True)
```

```python
drt = []
for d in map['director']:
  if ',' in d:
    d = [ int(df_drt.loc[df_drt['name']==i, 'id']) for i in d.split(',')]
    dc = {'id':d}
    drt.append(dc)
    continue
  d = [ i for i in df_drt.loc[df_drt['name']==d, 'id']]
  dc = {'id':d}
  drt.append(dc)

map_drt = pd.DataFrame(drt)
map['director'] = map_drt['id']
```

```python
pat = "{('id':\s)(\d{1,10})(.+)"
rep = r'\2'
map['genres'] = map['genres'].apply(lambda s: [re.sub(pat, rep, str(t)) for t in eval(s)])
```

```python
pat = "(.+)('id':\s)(\d{1,100})(.+)"
rep = r'\3'
map['cast'] = map['cast'].apply(lambda s: [re.sub(pat, rep, str(t)) for t in eval(s.replace('false', 'False').replace('null', 'None').replace('true', 'True'))])
```

```python
pat = '[A-Z]{2}'
map['production_countries'] = map['production_countries'].apply(lambda s: re.findall(pat, s))
```

```python
map
```

```python
"""
movie_genres,
movie_cast,
moive_countries
"""
movie_genres = { 'movie_id':[], 'genres_id':[]}
movie_cast = { 'movie_id':[], 'cast_id':[]}
movie_countries = { 'movie_id':[], 'countries_id':[]}
movie_drt = { 'movie_id':[], 'drt_id':[]}
for i in map['id']:
  for gen in map.loc[map['id']==i, 'genres']:
    for d in gen:
      movie_genres['movie_id'].append(i)
      movie_genres['genres_id'].append(d)
  for c in map.loc[map['id']==i, 'cast']:
    for d in c:
      movie_cast['movie_id'].append(i)
      movie_cast['cast_id'].append(d)
  for ct in map.loc[map['id']==i, 'production_countries']:
    for d in ct:
      movie_countries['movie_id'].append(i)
      movie_countries['countries_id'].append(d)
  for drt in map.loc[map['id']==i, 'director']:
    for d in drt:
      movie_drt['movie_id'].append(i)
      movie_drt['drt_id'].append(d)
```

```python
df_mov_gen = pd.DataFrame(movie_genres)
df_mov_cast = pd.DataFrame(movie_cast)
df_mov_ct = pd.DataFrame(movie_countries)
df_mov_drt = pd.DataFrame(movie_drt)
```

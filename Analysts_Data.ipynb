# อันดับจำนวนหนังที่เเสดงของนักแสดงในหนังแต่ละประเภท เเละได้คะเเนนเฉลี่ยเท่าไหร่

```python
sql =''' SELECT genres.name AS genres, caster.name AS caster, AVG(vote_average) AS vote_average, COUNT(*) AS number_of_movies
FROM movie, genres, movie_genres, caster, movie_caster
WHERE movie.id=movie_genres.movie_id
AND genres.id=movie_genres.genres_id
AND movie.id=movie_caster.movie_id
AND caster.id=movie_caster.cast_id
GROUP BY genres.name, caster.name
ORDER BY number_of_movies DESC
'''
df_cast_gen = pd.read_sql(sql, db)

cast_gen = {}
ls_genres = df_cast_gen['genres'].unique().tolist()

for gen in ls_genres:
  index = df_cast_gen.genres == gen
  cast_gen[gen] = df_cast_gen[index]
```

# อันดับจำนวนหนังที่กำกับของผู้กำกับในหนังแต่ละประเภท เเละได้คะเเนนเฉลี่ยเท่าไหร่

# นักแสดงคู่ใจของผู้กำกับ

```python
sql = ''' SELECT director.name AS director, caster.name AS caster, COUNT(*) AS number_of_movies
FROM director, caster, movie, movie_caster, movie_drt
WHERE movie_drt.drt_id = director.id
AND movie.id = movie_drt.movie_id
AND caster.id = movie_caster.cast_id
AND movie.id = movie_caster.movie_id
GROUP BY director, caster
ORDER BY number_of_movies DESC
'''

df_drt_cast = pd.read_sql(sql, db)
```

```python
df_drt_cast.head(10)
```

# อันดับจำนวนหนังที่ผู้กำกับเล่นด้วยตัวเอง

```python
index = df_drt_cast.apply(lambda x: x["director"]==x["caster"], axis=1)
df_drt_drt = df_drt_cast[index]
```

```python
df_drt_drt.head(10)
```

# อันดับของประเทศที่ผลิตหนังมากที่สุด

```python
sql = ''' SELECT countries.name AS country, COUNT(*) AS number_of_movies
FROM countries, movie, movie_ct
WHERE movie.id = movie_ct.movie_id
AND movie_ct.countries_id = countries.iso_3166_1
GROUP BY country
ORDER BY number_of_movies DESC
'''

df_movie_ct = pd.read_sql(sql, db)
df_movie_ct.head(10)
```

# จำนวนปีที่มีหนังมากที่สุด

```python
sql ='''SELECT release_year AS year ,COUNT(*) AS number_of_movies
FROM movie
GROUP BY release_year
ORDER BY number_of_movies DESC'''
df_year = pd.read_sql(sql, db)
```
